<tr>

</tr>
<tr>
<td>Central Composite<br>Design (CCD)</td>

<td>Minimal for<br>2nd-order model</td>
<td>Moderate (axial + factorial)</td>
<td>Second-order polynomial (quadratic)</td>
<td>Process optimization (3–5 factors); RSM</td>
<td>Efficient quadratic modeling; rotatability options; well-studied</td>
<td>Star points may exceed factor bounds; not space-filling</td>
<td>Blocking for star/factorial; randomization; center point replication</td>
</tr>
<tr>
<td>Box–Behnken<br>Design (BBD)</td>
<td>2k<sup>2</sup> + 2k + 1<br>(e.g., 13, 25, 41)</td>
<td>Main, quadratic, two-factor interactions</td>
<td>Minimal for<br>2nd-order model</td>
<td>Moderate (edge midpoints + center)</td>
<td>Second-order polynomial (quadratic)</td>
<td>Optimization (3–7 factors); avoiding extreme corners</td>

<td>Randomization; center point replication</td>

<td>Non-parametric; surrogate models (GP, neural nets)</td>
<td>Computer experiments; global sensitivity analysis</td>

<td>Random gaps possible; may cluster in high dimensions</td>
<td>Optimized LHS (maximin, orthogonal)</td>

### Table 2: Advanced and Specialized DOE Methods

<div style="overflow-x: auto;">
<table class="research-table-2">
<thead>
<tr>
<th>Design Method</th>
<th>Typical Run Count</th>
<th>Factors Estimable</th>
<th>Special Features</th>
<th>Model Type</th>
<th>Primary Application</th>
<th>Key Advantages</th>
<th>Key Limitations</th>
<th>Bias Mitigation</th>
</tr>
</thead>
<tbody>
<tr>
<td>D-Optimal Design</td>
<td>Variable<br>(optimized for specific N)</td>
<td>User-defined (main, interactions, quadratics)</td>
<td>Maximizes determinant of information matrix; handles constraints</td>
<td>Flexible<br>(polynomial, custom)</td>
<td>Constrained design spaces; irregular regions; mixture experiments</td>
<td>Handles complex constraints; mixed factor types; irregular boundaries; optimal for specific models</td>
<td>Model-dependent; requires candidate set; computationally intensive; no standard run formula</td>
<td>Model adequacy checking; augmentation strategies; sequential refinement</td>
</tr>
<tr>
<td>A-Optimal Design</td>
<td>Variable<br>(optimized)</td>
<td>User-defined</td>
<td>Minimizes average variance of parameter estimates</td>
<td>Flexible<br>(polynomial, custom)</td>
<td>Prediction-focused applications; parameter estimation precision</td>
<td>Minimizes average prediction variance; handles constraints</td>
<td>Model-dependent; different objectives than D-optimal; requires optimization algorithms</td>
<td>Model validation; cross-validation; sequential augmentation</td>
</tr>
<tr>
<td>Definitive Screening<br>Design (DSD)</td>
<td>2k + 1 for k factors<br>(e.g., 7, 13, 17)</td>
<td>Main effects, two-factor interactions, quadratics (with sparsity)</td>
<td>3-level design; no confounding of main effects; economical</td>
<td>Second-order with interaction selection</td>
<td>Combined screening and optimization; early product development</td>
<td>Estimates main, interactions, quadratics in one design; minimal runs; no aliasing of main effects</td>
<td>Requires effect sparsity assumption; 3 levels required; relatively new (limited long-term validation)</td>
<td>Randomization; augmentation if needed; effect heredity principles</td>
</tr>
<tr>
<td>Taguchi Orthogonal<br>Arrays</td>
<td>Standard arrays<br>(L8, L9, L12, L16, L18, L27, etc.)</td>
<td>Main effects primarily; some interactions</td>
<td>Inner (control) and outer (noise) array structure</td>
<td>Signal-to-noise ratio optimization</td>
<td>Robust parameter design; quality engineering; Six Sigma</td>
<td>Emphasizes robustness to noise; widely adopted in industry; standard arrays available</td>
<td>Heavy confounding; limited interaction estimation; complex analysis for interactions</td>
<td>Noise factor explicit consideration; confirmation runs; robust design philosophy</td>
</tr>
<tr>
<td>Mixture Designs<br>(Simplex-Lattice)</td>
<td>C(q+m-1, m) for<br>q levels, m components</td>
<td>Mixture effects; blending interactions</td>
<td>Factors are proportions summing to 1 (or 100%)</td>
<td>Scheffe polynomials (mixture models)</td>
<td>Formulation (food, pharma, chemicals, cosmetics); blending</td>
<td>Handles compositional constraints naturally; specialized models for mixtures</td>
<td>Constrained to simplex space; cannot test zero proportions; specialized analysis</td>
<td>Pseudo-components for constrained regions; replication of interior points</td>
</tr>
<tr>
<td>Mixture Designs<br>(Simplex-Centroid)</td>
<td>2<sup>m</sup> - 1<br>for m components</td>
<td>Mixture effects including centroids</td>
<td>Includes all vertices, edge centers, and overall centroid</td>
<td>Scheffe polynomials</td>
<td>Formulation with emphasis on blend centers</td>
<td>More points at interior; better for nonlinear blending; includes overall center</td>
<td>More runs than simplex-lattice; constrained space</td>
<td>Interior point replication; pseudo-components</td>
</tr>
<tr>
<td>Split-Plot Designs</td>
<td>Depends on whole-plot and sub-plot structure</td>
<td>Main effects and interactions (with separate error terms)</td>
<td>Two levels of randomization; hard-to-change and easy-to-change factors</td>
<td>Mixed-effects models</td>
<td>Experiments with hard-to-change factors (temperature, batch, equipment)</td>
<td>Reflects practical constraints; reduces cost of changing hard factors; realistic industrial model</td>
<td>Complex analysis (two error terms); restricted randomization; requires mixed-model expertise</td>
<td>Proper randomization at both levels; blocking; REML analysis</td>
</tr>
<tr>
<td>Sobol Sequences</td>
<td>Flexible<br>(deterministic)</td>
<td>Model-free</td>
<td>Low-discrepancy quasi-random sequence; superior uniformity</td>
<td>Non-parametric; surrogate models</td>
<td>Global sensitivity analysis; computer experiments; Monte Carlo methods</td>
<td>Excellent multi-dimensional uniformity; deterministic; reproducible; sequentially extendable</td>
<td>Dimension-dependent performance; requires initialization; more complex than Halton</td>
<td>Deterministic uniformity; scrambling variants for improved properties</td>
</tr>
<tr>
<td>Uniform Design</td>
<td>U<sub>n</sub>(q<sup>m</sup>):<br>n runs, m factors, q levels</td>
<td>Main effects; model-free exploration</td>
<td>Number-theoretic uniformity; discrepancy minimization</td>
<td>Flexible/<br>non-parametric</td>
<td>Many-factor screening; non-polynomial relationships; complex systems</td>
<td>Handles many factors with few runs; good uniformity; flexible modeling</td>
<td>Specialized tables needed; less familiar in Western industry; complex construction</td>
<td>Uniformity ensures balanced coverage; validation experiments</td>
</tr>
<tr>
<td>Adaptive/Sequential<br>Designs</td>
<td>Sequential accumulation (iterative)</td>
<td>Depends on learning algorithm</td>
<td>Next experiment selected based on prior results; active learning</td>
<td>Bayesian, Gaussian process, neural networks</td>
<td>Expensive experiments; optimization under uncertainty; AI-integrated DOE</td>
<td>Efficient learning; adapts to data; integrates prior knowledge; optimal resource allocation</td>
<td>Requires sophisticated algorithms; sequential (not parallelizable); model-dependent</td>
<td>Bayesian updating; exploration-exploitation balance; cross-validation</td>
</tr>
</tbody>
</table>
</div>

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

# Define a quadratic response surface function
def response_surface(x1, x2):
    """Simulated quadratic response function"""
    return 80 + 5*x1 + 3*x2 - 2*x1**2 - 1.5*x2**2 + 1.5*x1*x2

# Create Central Composite Design points (CCD) for 2 factors
# Factorial points (corners)
factorial_points = np.array([
    [-1, -1], [1, -1], [-1, 1], [1, 1]
])

# Axial/star points (alpha = 1.414 for rotatability)
alpha = np.sqrt(2)
axial_points = np.array([
    [-alpha, 0], [alpha, 0], [0, -alpha], [0, alpha]
])

# Center points
center_points = np.array([[0, 0]])

# Combine all CCD points
ccd_points = np.vstack([factorial_points, axial_points, center_points])

# Create figure with two subplots
fig = plt.figure(figsize=(16, 6))

# Subplot 1: 3D Response Surface
ax1 = fig.add_subplot(1, 2, 1, projection='3d')

# Create mesh grid for surface
x1_range = np.linspace(-2, 2, 50)
x2_range = np.linspace(-2, 2, 50)
X1, X2 = np.meshgrid(x1_range, x2_range)
Y = response_surface(X1, X2)

# Plot surface
surf = ax1.plot_surface(X1, X2, Y, cmap=cm.viridis, alpha=0.7, 
                        edgecolor='none', linewidth=0, antialiased=True)

# Plot CCD design points
response_at_points = response_surface(ccd_points[:, 0], ccd_points[:, 1])

# Factorial points (red)
ax1.scatter(factorial_points[:, 0], factorial_points[:, 1], 
           response_surface(factorial_points[:, 0], factorial_points[:, 1]),
           c='red', marker='o', s=150, edgecolors='black', linewidths=2, 
           label='Factorial Points', zorder=5)

# Axial points (blue)
ax1.scatter(axial_points[:, 0], axial_points[:, 1], 
           response_surface(axial_points[:, 0], axial_points[:, 1]),
           c='blue', marker='s', s=150, edgecolors='black', linewidths=2, 
           label='Axial Points', zorder=5)

# Center point (green)
ax1.scatter(center_points[:, 0], center_points[:, 1], 
           response_surface(center_points[:, 0], center_points[:, 1]),
           c='green', marker='^', s=200, edgecolors='black', linewidths=2, 
           label='Center Point', zorder=5)

ax1.set_xlabel('Factor X₁', fontsize=11, fontweight='bold')
ax1.set_ylabel('Factor X₂', fontsize=11, fontweight='bold')
ax1.set_zlabel('Response Y', fontsize=11, fontweight='bold')
ax1.set_title('3D Response Surface with CCD Points', 
              fontsize=13, fontweight='bold', pad=15)
ax1.legend(loc='upper left', fontsize=9)
ax1.view_init(elev=25, azim=45)

# Add colorbar
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=10, label='Response Value')

# Subplot 2: Contour Plot with Design Points
ax2 = plt.subplot(1, 2, 2)

# Create contour plot
contour = ax2.contourf(X1, X2, Y, levels=15, cmap=cm.viridis, alpha=0.8)
contour_lines = ax2.contour(X1, X2, Y, levels=15, colors='black', 
                            linewidths=0.5, alpha=0.4)
ax2.clabel(contour_lines, inline=True, fontsize=8, fmt='%0.1f')

# Plot CCD design points
ax2.scatter(factorial_points[:, 0], factorial_points[:, 1], 
           c='red', marker='o', s=150, edgecolors='black', linewidths=2, 
           label='Factorial Points', zorder=5)
ax2.scatter(axial_points[:, 0], axial_points[:, 1], 
           c='blue', marker='s', s=150, edgecolors='black', linewidths=2, 
           label='Axial Points', zorder=5)
ax2.scatter(center_points[:, 0], center_points[:, 1], 
           c='green', marker='^', s=200, edgecolors='black', linewidths=2, 
           label='Center Point', zorder=5)

# Mark the optimum region
optimum_idx = np.unravel_index(np.argmax(Y), Y.shape)
opt_x1 = X1[optimum_idx]
opt_x2 = X2[optimum_idx]
ax2.plot(opt_x1, opt_x2, 'r*', markersize=20, markeredgecolor='yellow', 
         markeredgewidth=2, label='Optimum Region', zorder=6)

ax2.set_xlabel('Factor X₁', fontsize=11, fontweight='bold')
ax2.set_ylabel('Factor X₂', fontsize=11, fontweight='bold')
ax2.set_title('Contour Plot with CCD Design Points\nand Optimum Region', 
              fontsize=13, fontweight='bold')
ax2.legend(loc='upper right', fontsize=9)
ax2.grid(True, alpha=0.3, linestyle='--')
ax2.set_xlim([-2, 2])
ax2.set_ylim([-2, 2])
ax2.axhline(y=0, color='gray', linestyle='-', linewidth=0.5, alpha=0.5)
ax2.axvline(x=0, color='gray', linestyle='-', linewidth=0.5, alpha=0.5)

# Add colorbar
cbar = fig.colorbar(contour, ax=ax2, label='Response Value')

plt.tight_layout()
plt.savefig('Figure2_Response_Surface_CCD.png', dpi=300, bbox_inches='tight')
plt.show()

print("Figure 2 saved as 'Figure2_Response_Surface_CCD.png'")
print(f"\nCCD Design Points (2 factors, α={alpha:.3f}):")
print("Factorial points:", factorial_points.shape[0])
print("Axial points:", axial_points.shape[0])
print("Center points:", center_points.shape[0])
print(f"Total runs: {ccd_points.shape[0]}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Simulated screening experiment results
# 7 factors screened with standardized effects
factors = ['Temperature', 'Pressure', 'Catalyst Conc.', 'Reaction Time', 
           'pH Level', 'Stirring Speed', 'Feed Rate']

# Simulated standardized effects (absolute values)
effects = np.array([4.8, 3.2, 2.9, 1.8, 1.2, 0.8, 0.5])

# Significance threshold (t-value for α=0.05, df=8)
t_critical = 2.306  # for 95% confidence
significance_line = t_critical

# Sort factors by effect magnitude (descending)
sorted_indices = np.argsort(effects)[::-1]
sorted_factors = [factors[i] for i in sorted_indices]
sorted_effects = effects[sorted_indices]

# Determine which effects are significant
significant = sorted_effects > significance_line

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# ============ Subplot 1: Pareto Chart ============
colors = ['#d32f2f' if sig else '#757575' for sig in significant]
bars = ax1.barh(range(len(sorted_factors)), sorted_effects, color=colors, 
                edgecolor='black', linewidth=1.5, alpha=0.85)

# Add significance threshold line
ax1.axvline(x=significance_line, color='blue', linestyle='--', linewidth=2.5, 
            label=f'Significance Threshold (α=0.05)\nt = {significance_line:.2f}')

# Customize plot
ax1.set_yticks(range(len(sorted_factors)))
ax1.set_yticklabels(sorted_factors, fontsize=11)
ax1.set_xlabel('Standardized Effect (Absolute Value)', fontsize=12, fontweight='bold')
ax1.set_title('Pareto Chart: Factor Screening Results\n(Ranked by Effect Magnitude)', 
              fontsize=13, fontweight='bold', pad=15)
ax1.legend(loc='lower right', fontsize=10)
ax1.grid(axis='x', alpha=0.3, linestyle='--')
ax1.set_xlim([0, max(sorted_effects) * 1.15])

# Add value labels on bars
for i, (bar, effect) in enumerate(zip(bars, sorted_effects)):
    ax1.text(effect + 0.15, i, f'{effect:.2f}', 
            va='center', fontsize=10, fontweight='bold')

# Add legend for colors
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='#d32f2f', edgecolor='black', label='Significant Factors'),
    Patch(facecolor='#757575', edgecolor='black', label='Non-significant Factors')
]
ax1.legend(handles=legend_elements, loc='lower right', fontsize=10)

# ============ Subplot 2: Normal Probability Plot ============
# Simulate signed effects (with positive and negative signs)
signed_effects = np.array([4.8, -3.2, 2.9, -1.8, 1.2, 0.8, -0.5])
sorted_signed = np.sort(signed_effects)

# Calculate theoretical quantiles
n = len(signed_effects)
quantiles = stats.norm.ppf(np.arange(1, n + 1) / (n + 1))

# Identify significant effects (those far from the line)
abs_effects = np.abs(signed_effects)
significant_mask = abs_effects > significance_line

# Plot
ax2.scatter(quantiles[~significant_mask], sorted_signed[~significant_mask], 
           s=100, c='gray', marker='o', edgecolors='black', linewidths=1.5,
           label='Non-significant Effects', zorder=3)
ax2.scatter(quantiles[significant_mask], sorted_signed[significant_mask], 
           s=150, c='red', marker='o', edgecolors='black', linewidths=1.5,
           label='Significant Effects', zorder=4)

# Fit a line through non-significant effects (those near zero)
small_effects_mask = abs_effects < 2.0
if np.sum(small_effects_mask) > 1:
    slope, intercept = np.polyfit(quantiles[small_effects_mask], 
                                   sorted_signed[small_effects_mask], 1)
    line_x = np.array([quantiles.min(), quantiles.max()])
    line_y = slope * line_x + intercept
    ax2.plot(line_x, line_y, 'b--', linewidth=2, label='Normal Line', zorder=2)

# Add labels to significant points
for i, (q, eff) in enumerate(zip(quantiles, sorted_signed)):
    if abs(eff) > significance_line:
        factor_name = factors[np.where(signed_effects == eff)[0][0]]
        ax2.annotate(factor_name, xy=(q, eff), xytext=(10, 10),
                    textcoords='offset points', fontsize=9,
                    bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7),
                    arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0', lw=1))

ax2.set_xlabel('Theoretical Quantiles (Standard Normal)', fontsize=11, fontweight='bold')
ax2.set_ylabel('Standardized Effects', fontsize=11, fontweight='bold')
ax2.set_title('Normal Probability Plot of Effects\n(Significant factors deviate from line)', 
              fontsize=13, fontweight='bold', pad=15)
ax2.legend(loc='upper left', fontsize=10)
ax2.grid(True, alpha=0.3, linestyle='--')
ax2.axhline(y=0, color='black', linestyle='-', linewidth=0.8, alpha=0.5)
ax2.axhline(y=significance_line, color='red', linestyle=':', linewidth=1.5, alpha=0.7)
ax2.axhline(y=-significance_line, color='red', linestyle=':', linewidth=1.5, alpha=0.7)

plt.tight_layout()
plt.savefig('Figure3_Pareto_Chart_Screening.png', dpi=300, bbox_inches='tight')
plt.show()

print("Figure 3 saved as 'Figure3_Pareto_Chart_Screening.png'")
print(f"\nScreening Results Summary:")
print(f"Total factors screened: {len(factors)}")
print(f"Significant factors (α=0.05): {np.sum(significant)}")
print(f"Significant factors: {', '.join([sorted_factors[i] for i in range(len(sorted_factors)) if significant[i]])}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon, Circle
from scipy.spatial import ConvexHull
from sklearn.preprocessing import PolynomialFeatures
from numpy.linalg import det

# Set random seed
np.random.seed(42)

# ============ Define Constrained Region ============
# Create an irregular feasible region (e.g., process constraints)
# Region defined by: x1 + x2 <= 1.5, x1 >= 0.2, x2 >= 0.2, x1 <= 1, x2 <= 1
# Plus a circular "forbidden zone"

def is_in_feasible_region(x1, x2):
    """Check if point is in feasible region"""
    # Basic bounds
    if x1 < 0 or x1 > 1 or x2 < 0 or x2 > 1:
        return False
    # Linear constraint
    if x1 + x2 > 1.5:
        return False
    # Minimum values
    if x1 < 0.2 or x2 < 0.2:
        return False
    # Forbidden circular zone (e.g., unsafe operating region)
    if (x1 - 0.6)**2 + (x2 - 0.6)**2 < 0.08:
        return False
    return True

# Generate candidate points
n_candidates = 2000
candidates = []
while len(candidates) < n_candidates:
    x1, x2 = np.random.uniform(0, 1, 2)
    if is_in_feasible_region(x1, x2):
        candidates.append([x1, x2])
candidates = np.array(candidates)

# ============ Generate D-Optimal Design ============
def calculate_d_criterion(X_design, poly_degree=2):
    """Calculate D-optimality criterion (determinant of X'X)"""
    poly = PolynomialFeatures(degree=poly_degree, include_bias=True)
    X_model = poly.fit_transform(X_design)
    XTX = X_model.T @ X_model
    try:
        return det(XTX)
    except:
        return 0

# Greedy D-optimal selection
n_design_points = 15
d_optimal_points = []

# Start with best single point
best_det = -np.inf
for point in candidates[:200]:  # Check subset for speed
    det_val = calculate_d_criterion(point.reshape(1, -1))
    if det_val > best_det:
        best_det = det_val
        best_point = point

d_optimal_points.append(best_point)

# Iteratively add points that maximize determinant
for i in range(n_design_points - 1):
    best_det = -np.inf
    best_point = None
    
    current_design = np.array(d_optimal_points)
    
    # Sample candidates for computational efficiency
    sample_indices = np.random.choice(len(candidates), min(300, len(candidates)), replace=False)
    
    for idx in sample_indices:
        candidate = candidates[idx]
        # Skip if already selected
        if any(np.allclose(candidate, p) for p in d_optimal_points):
            continue
        
        # Test this candidate
        test_design = np.vstack([current_design, candidate])
        det_val = calculate_d_criterion(test_design)
        
        if det_val > best_det:
            best_det = det_val
            best_point = candidate
    
    if best_point is not None:
        d_optimal_points.append(best_point)

d_optimal_points = np.array(d_optimal_points)

# ============ Generate CCD for Comparison (ignoring constraints) ============
# Standard CCD points (will show some outside feasible region)
alpha = np.sqrt(2)
ccd_points = np.array([
    [0.2, 0.2], [0.8, 0.2], [0.2, 0.8], [0.8, 0.8],  # Factorial (scaled)
    [0.5 - 0.3*alpha, 0.5], [0.5 + 0.3*alpha, 0.5],  # Axial
    [0.5, 0.5 - 0.3*alpha], [0.5, 0.5 + 0.3*alpha],
    [0.5, 0.5], [0.5, 0.5], [0.5, 0.5]  # Center points
])

# ============ Create Figure ============
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# ============ Subplot 1: CCD (Classical Design - ignores constraints) ============
ax1 = axes[0]

# Plot feasible region
x1_grid = np.linspace(0, 1, 300)
x2_grid = np.linspace(0, 1, 300)
X1_grid, X2_grid = np.meshgrid(x1_grid, x2_grid)
feasible_mask = np.array([[is_in_feasible_region(x1, x2) 
                          for x1 in x1_grid] for x2 in x2_grid])

ax1.contourf(X1_grid, X2_grid, feasible_mask, levels=[0.5, 1.5], 
            colors=['lightgreen'], alpha=0.3)
ax1.contour(X1_grid, X2_grid, feasible_mask, levels=[0.5], 
           colors=['darkgreen'], linewidths=2)

# Plot forbidden zone
forbidden_circle = Circle((0.6, 0.6), 0.283, color='red', alpha=0.2, 
                         edgecolor='darkred', linewidth=2, linestyle='--',
                         label='Forbidden Zone')
ax1.add_patch(forbidden_circle)

# Plot CCD points
ccd_feasible = np.array([p for p in ccd_points if is_in_feasible_region(p[0], p[1])])
ccd_infeasible = np.array([p for p in ccd_points if not is_in_feasible_region(p[0], p[1])])

if len(ccd_feasible) > 0:
    ax1.scatter(ccd_feasible[:, 0], ccd_feasible[:, 1], c='blue', marker='o', 
               s=150, edgecolors='black', linewidths=2, alpha=0.8, 
               label=f'CCD Points (Feasible: {len(ccd_feasible)})', zorder=5)

if len(ccd_infeasible) > 0:
    ax1.scatter(ccd_infeasible[:, 0], ccd_infeasible[:, 1], c='gray', marker='x', 
               s=200, linewidths=3, alpha=0.8, 
               label=f'CCD Points (Infeasible: {len(ccd_infeasible)})', zorder=5)

ax1.set_xlim([0, 1])
ax1.set_ylim([0, 1])
ax1.set_xlabel('Factor X₁', fontsize=12, fontweight='bold')
ax1.set_ylabel('Factor X₂', fontsize=12, fontweight='bold')
ax1.set_title('Classical Central Composite Design (CCD)\n(Ignores Constraints - Some Points Infeasible)', 
             fontsize=12, fontweight='bold')
ax1.legend(loc='upper right', fontsize=9)
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.set_aspect('equal')

# Add annotation
ax1.text(0.5, 0.05, 'Feasible Region (Green)', ha='center', fontsize=10,
        bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.7))

# ============ Subplot 2: D-Optimal Design (adapts to constraints) ============
ax2 = axes[1]

# Plot feasible region
ax2.contourf(X1_grid, X2_grid, feasible_mask, levels=[0.5, 1.5], 
            colors=['lightgreen'], alpha=0.3)
ax2.contour(X1_grid, X2_grid, feasible_mask, levels=[0.5], 
           colors=['darkgreen'], linewidths=2, label='Feasible Boundary')

# Plot forbidden zone
forbidden_circle2 = Circle((0.6, 0.6), 0.283, color='red', alpha=0.2, 
                          edgecolor='darkred', linewidth=2, linestyle='--',
                          label='Forbidden Zone')
ax2.add_patch(forbidden_circle2)

# Plot candidate points (sample)
sample_candidates = candidates[::10]  # Show every 10th for clarity
ax2.scatter(sample_candidates[:, 0], sample_candidates[:, 1], 
           c='lightgray', marker='.', s=10, alpha=0.3, label='Candidate Points')

# Plot D-optimal points
ax2.scatter(d_optimal_points[:, 0], d_optimal_points[:, 1], c='red', marker='o', 
           s=150, edgecolors='black', linewidths=2, alpha=0.9, 
           label=f'D-Optimal Points (n={len(d_optimal_points)})', zorder=5)

# Number the D-optimal points
for i, point in enumerate(d_optimal_points):
    ax2.annotate(f'{i+1}', xy=(point[0], point[1]), xytext=(5, 5),
                textcoords='offset points', fontsize=8, fontweight='bold',
                color='white', bbox=dict(boxstyle='circle', facecolor='red', 
                                        edgecolor='black', linewidth=1))

ax2.set_xlim([0, 1])
ax2.set_ylim([0, 1])
ax2.set_xlabel('Factor X₁', fontsize=12, fontweight='bold')
ax2.set_ylabel('Factor X₂', fontsize=12, fontweight='bold')
ax2.set_title('D-Optimal Design\n(Adapts to Constrained Region - All Points Feasible)', 
             fontsize=12, fontweight='bold')
ax2.legend(loc='upper right', fontsize=9)
ax2.grid(True, alpha=0.3, linestyle='--')
ax2.set_aspect('equal')

# Add annotation
ax2.text(0.5, 0.05, 'Optimized for Quadratic Model', ha='center', fontsize=10,
        bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))

# Main title
fig.suptitle('Comparison: Classical CCD vs. D-Optimal Design in Constrained Space', 
            fontsize=14, fontweight='bold', y=0.98)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('Figure6_D_Optimal_Constrained.png', dpi=300, bbox_inches='tight')
plt.show()

print("Figure 6 saved as 'Figure6_D_Optimal_Constrained.png'")
print(f"\nD-Optimal Design Summary:")
print(f"  Total candidate points: {len(candidates)}")
print(f"  Selected D-optimal points: {len(d_optimal_points)}")
print(f"  All D-optimal points are feasible: {all(is_in_feasible_region(p[0], p[1]) for p in d_optimal_points)}")
print(f"\nCCD Comparison:")
print(f"  Total CCD points: {len(ccd_points)}")
print(f"  Feasible CCD points: {len(ccd_feasible)}")
print(f"  Infeasible CCD points: {len(ccd_infeasible)}")
print(f"  D-Optimal advantage: Respects all constraints while maximizing information")
```
